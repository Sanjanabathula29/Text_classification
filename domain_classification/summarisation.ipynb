```python
from transformers import pipeline
```

```python
text = """
Lorem ipsum dolor sit amet, consectetur adipiscing elit. Nullam tincidunt euismod ex, a consequat velit semper ac.
Vestibulum ante ipsum primis in faucibus orci luctus et ultrices posuere cubilia Curae; Nulla facilisi. Duis dapibus
purus mi, eget suscipit tellus eleifend at. Mauris sit amet pulvinar nunc. Donec interdum velit in odio ultricies
elementum. Proin id est elit. Pellentesque habitant morbi tristique senectus et netus et malesuada fames ac turpis
egestas. Sed pharetra leo vel mauris consequat, eu finibus odio eleifend. Quisque id fringilla libero. Ut non
eleifend erat, eu pretium ex. Vivamus eu eleifend mi, a gravida lectus. Aliquam in fringilla erat. Sed condimentum
lacinia justo et ullamcorper. Vestibulum ante ipsum primis in faucibus orci luctus et ultrices posuere cubilia Curae;
"""
```

```python
summarizer = pipeline("summarization")
```

```python
summary = summarizer(text, max_length=100, min_length=30, do_sample=False)
```

```python
print(summary[0]['summary_text'])
```
